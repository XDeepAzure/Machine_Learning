```python
import os
import pandas as pd
from typing import List
from math import log2
```

```python
def getData(file_path:str) ->List[List[any]]:
    '''
    去掉第一行头
    '''
    with open(file_path, encoding='utf-8') as f:
        data = f.readlines()
    head = data[0].rstrip().split(' ')
    body, data = [[] for _ in range(len(head))], data[1:]
    for line in data:
        if line == '\n':
            break
        for idx, item in enumerate(line.rstrip().split(' ')):
            body[idx].append(item)
    return (head, body)
```

```python
head, body  = getData('./贷款申请训练集.txt')
head, body
```

```python
dic = {head[i]:body[i] for i in range(len(head))}
dic
```

```python
def splitSetInParam(D, param):
    #在集合D中以param进行划分，一个param值划分一个集合
    catalog = tuple(set(D[param]))     #属性p有多少种类，分别是啥
    set_lst = []                            #对p中每种catalog划分集合
    for c in catalog:
        set_lst.append(D.loc[D[param] == c])
    return set_lst
```

```python
def informEntropy(D, type='类别'):
    #计算信息熵，type是类别即数据集最终类别
    catalog, nums_lst = tuple(set(D[type])), []
    for c in catalog:                       #统计分类中，每个类别所占个数
        nums_lst.append(len(D[type] == c))
    sum_num, ent = sum(nums_lst), 0         #此集合中样本总数，ent的值
    for n in nums_lst:
        p_k = n / sum_num
        ent -= p_k * log2(p_k)
    return ent
    pass
informEntropy(train_data)                   
```

```python
def informGain(D, type):
    #计算D以type划分的信息增益
    num_D = len(D)
    catalog = set(D[type])
    sub_set = [D.loc[D[type] == c] for c in catalog]        #按照type将D进行划分
    gain = informEntropy(D)
    for D_v in sub_set:
        gain -= (len(D_v) / num_D) * informEntropy(D_v)
    return gain
    pass
```
